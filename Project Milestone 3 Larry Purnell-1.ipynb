```python
vote2024 = pd.read_csv("vote01_2024_e.csv")

# This code cell loads vote01_2024.csv 
```

```python
vote2024.keys()
```

```python
vote2024_fixed = vote2024.copy()
```

```python
vote2024_fixed = vote2024.rename(columns={
    'sex': 'Sex',
    'age': 'Age_Group',
    'tot_pop': 'Total_Population',
    'us_cit_tot_pop': 'Citizen_Population',
    
    'us_cit_rep_reg_num': 'Registered_Number',
    'us_cit_rep_reg_pct': 'Registered_Percent',
    
    'us_cit_rep_nr_num': 'Not_Registered_Number',
    'us_cit_rep_nr_pct': 'Not_Registered_Percent',
    
    'us_cit_nr_num': 'No_Response_Registration_Number',
    'us_cit_nr_pct': 'No_Response_Registration_Percent',
    
    'us_cit_vote_yes_num': 'Voted_Number',
    'us_cit_vote_yes_pct': 'Voted_Percent',
    
    'us_cit_vote_no_num': 'Did_Not_Vote_Number',
    'us_cit_vote_no_pct': 'Did_Not_Vote_Percent',
    
    'us_cit_vote_nr_num': 'No_Response_Vote_Number',
    'us_cit_vote_nr_pct': 'No_Response_Vote_Percent',
    
    'us_tot_rep_reg_pct': 'Overall_Registration_Percent',
    'us_tot_rep_vote_yes_pct': 'Overall_Vote_Percent'
})


# renamed columns to accurate repsentation of the updated cloumns in the csv file. 
```

```python
vote2024_fixed.head(10)
```

```python
vote2024_fixed = pd.DataFrame(vote2024_fixed)
```

```python
young_voters = vote2024_fixed[vote2024_fixed['Age_Group'] == '18-29']
```

```python
print(vote2024_fixed['Age_Group'].unique())
```

```python
young_voters_melted = young_voters.melt(
    value_vars=['Registered_Number', 'Not_Registered_Number'],
    var_name='Registration_Status',
    value_name='Count'
)
```

```python
percent_melted = young_voters.melt(
    value_vars=['Overall_Registration_Percent', 'Overall_Vote_Percent'],
    var_name='Metric',
    value_name='Percent'
)

percent_melted['Metric'] = percent_melted['Metric'].replace({
    'Overall_Registration_Percent': 'Registered (%)',
    'Overall_Vote_Percent': 'Voted (%)'
})

fig3 = px.bar(
    percent_melted,
    x='Metric',
    y='Percent',
    title='Registration vs Voting Percentage (Ages 18–29)',
    text='Percent'
)

fig3.show()
```

```python
young_voters_melted = young_voters.melt(
    value_vars=['Voted_Number', 'Did_Not_Vote_Number'],
    var_name='Vote_Status',
    value_name='Count'
)
```

```python
fig3 = px.pie(
    young_voters_melted,
    names='Vote_Status',
    values='Count',
    title='Voted vs Did Not Vote (Ages 18–29)'

)

fig3.show()
```

```python
fig4 = px.bar(
    young_voters_melted,
    x='Vote_Status',
    y='Count',
    title='Voted vs Did Not Vote (Ages 18–29)',
    text='Count'
)

fig4.show()
```

```python
fig6 = px.scatter(
    vote2024_fixed,
    x ="Not_Registered_Number",
    y ="Registered_Number",
    title="Registered vs Not Registered Voters"
)

fig6.show()
```

```python
sex_summary = (
    vote2024_fixed
    .groupby('Sex')[['Registered_Percent', 'Voted_Percent']]
    .mean()
    .reset_index()
)

fig = px.bar(
    sex_summary,
    x='Sex',
    y=['Registered_Percent', 'Voted_Percent'],
    barmode='group',
    title='Registration vs Voting by Sex',
    labels={'value': 'Percent', 'variable': 'Metric'}
)

fig.show()
```
