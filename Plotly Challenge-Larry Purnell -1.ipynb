```python

import pandas as pd
import plotly
import plotly.express as px
import plotly.io as pio

#imports plotly, plotyIO, pandas, and ploty express. 
```

```python
df = pd.read_csv("daily_food_delivery_orders.csv")

#reads the csv file daily food delivery. 
```

```python
restaurant_counts = df["restaurant_type"].value_counts().reset_index()
restaurant_counts.columns = ["restaurant_type", "count"]
```

```python
fig1 = px.bar(
    restaurant_counts,
    x="restaurant_type",
    y="count",
    title="Frequency of Restaurant Type",
    labels={
        "restaurant_type": "Restaurant Type",
        "count": "Count"
    }
)
# Created a bar graph with the x as resteruant type, and y as Count. 
fig1.show()
```

```python
df_line = df.iloc[0:20].copy()

df_line["order_date"] = pd.to_datetime(df_line["order_date"])

df_line = df_line.sort_values("order_date")

#Slects rows 0-20, line two of code sorts by order_date
```

```python
fig2 = px.line(
    df_line,
    x="order_date",
    y="order_value",
    color="order_status",
    markers=True,
    title="Order Value Over Time"
)

fig2.update_traces(
    marker=dict(
        symbol="star-diamond",
        color="yellow",
        size=12
    )
)


fig2.show()

# creates line chart grpah with markers = True, title Order Caule over time, and X and y vaule.  
# the update_traces updates the marker to yellow star diamond, 
```

```python
df_random = df.sample(n=50, random_state=1)
```

```python
fig3 = px.scatter(
    df_random,
    x="delivery_time_minutes",
    y="delivery_distance_km",
    color="delivery_partner_rating",   # color coded ratings
    symbol="order_status",             # different shapes per order status
    size="delivery_distance_km",       # changes plot size
    title="Delivery Time",
    labels={
        "delivery_time_minutes": "Delivery Time (Minutes)",
        "delivery_distance_km": "Delivery Distance (KM)",
        "delivery_partner_rating": "Partner Rating",
        "order_status": "Order Status"
    },
    width=1000,
    height=500
)

fig3.show()
```

```python
df["age_group"] = pd.cut(
    df["customer_age"],
    bins=[18,25,35,45,60,80],
    labels=["18-25", "26-35", "36-45", "46-60", "60+"]
)
```

```python
fig = px.histogram(
    df_random,
    x="age_group",
    title="Frequency Distribution of Customer Age Groups",
    labels={
        "age_group": "Age Group",
        "count": "Frequency"
    },
    width=900,
    height=600
)

fig.show()

```
