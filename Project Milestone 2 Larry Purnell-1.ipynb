```python
vote2024_fixed = vote2024.copy()
```

```python
vote2024_fixed = vote2024.rename(columns={
    'sex': 'Sex',
    'age': 'Age_Group',
    'tot_pop': 'Total_Population',
    'us_cit_tot_pop': 'Citizen_Population',
    
    'us_cit_rep_reg_num': 'Registered_Number',
    'us_cit_rep_reg_pct': 'Registered_Percent',
    
    'us_cit_rep_nr_num': 'Not_Registered_Number',
    'us_cit_rep_nr_pct': 'Not_Registered_Percent',
    
    'us_cit_nr_num': 'No_Response_Registration_Number',
    'us_cit_nr_pct': 'No_Response_Registration_Percent',
    
    'us_cit_vote_yes_num': 'Voted_Number',
    'us_cit_vote_yes_pct': 'Voted_Percent',
    
    'us_cit_vote_no_num': 'Did_Not_Vote_Number',
    'us_cit_vote_no_pct': 'Did_Not_Vote_Percent',
    
    'us_cit_vote_nr_num': 'No_Response_Vote_Number',
    'us_cit_vote_nr_pct': 'No_Response_Vote_Percent',
    
    'us_tot_rep_reg_pct': 'Overall_Registration_Percent',
    'us_tot_rep_vote_yes_pct': 'Overall_Vote_Percent'
})


# renamed columns to accurate repsentation of the updated cloumns in the csv file. 
```

```python
vote2024_fixed.head(10)
```

```python
sns.scatterplot(data=vote2024_fixed, x="Registered_Number", y="Not_Registered_Number") 

plt.title("Registered vs Not Registered Voters")
plt.xlabel("Registered (Voted)")
plt.ylabel("Not Registered (Not Voted)")
plt.show()

# making a anaylisis to compare the Registered_number and Not Registered_number
```

```python

sns.histplot(data=vote2024_fixed,x="Registered_Number", y="Not_Registered_Number")
plt.show()
```

```python
# Remove commas and convert to numbers, Python can read the percentage as a number removes the 

vote2024_fixed["Registered_Number"] = (
    vote2024_fixed["Registered_Number"]
    .astype(str)
    .str.replace(",", "")
    .astype(float)
)

vote2024_fixed["Not_Registered_Number"] = (
    vote2024_fixed["Not_Registered_Number"]
    .astype(str)
    .str.replace(",", "")
    .astype(float)
)
```

```python


plt.figure 

plt.scatter(vote2024_fixed["Registered_Number"],vote2024_fixed["Not_Registered_Number"],color="#2E86AB",alpha=0.7)

plt.title("Registered vs Not Registered Voters", fontsize=14)
plt.xlabel("Registered Voters")
plt.ylabel("Not Registered Voters")

plt.grid(alpha = 1)
plt.tight_layout()
plt.show()
```
