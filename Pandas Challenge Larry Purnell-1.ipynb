What does the "write/render/evaluate" loop provide for creators of interactive data visualizations?

 The write/render/evaluate aides in the process of creating digital visualizations. Each step of the process allows the coder to change how the data is communicated. By using this process, you can manipulate digital visualizations by changing the mapping rules, seeing the chart's visual components, and evaluating where to change the visualization to fit your needs. 

Why are dynamic, interactive data visualizations useful when communicating data?

Dynamic and interactive data visualizations are more accessible to a variety of audiences. While static data visualizations cannot exactly get the point across and are used repeatedly to apply the same message, this makes the static visualization redundant. Dynamic data visualizations are used to create engagement with the viewer.
